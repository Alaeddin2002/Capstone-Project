```r
#1
bloodtypes <-  c("O" , "A", "B" , "AB")
bloodprobs <- c ( .45 , .40 , .11 , .04 )
bloodtypepop <- sample( bloodtypes , 10000 , prob=bloodprobs ,
replace=TRUE)
```

```r
table ( bloodtypepop )
table ( bloodtypepop )/10000

#We can see from table 1 that from 10000 people, The most common blood types are at the same order as the probability of them.
#From table2 we can see that the relative frequency is also ordered as the probability of each blood type. So the most common blood type to have is O which has the highest probability.
```

```r
simdataO <- bloodtypepop== "O"
simdataA <- bloodtypepop == "A"
simdataB <- bloodtypepop == "B"
simdataAB <- bloodtypepop == "AB"
```

```r
mean( simdataO )
mean( simdataA )
mean( simdataB )
mean( simdataAB )

# The mean is exactly the same as the relative frequency, as the mean is calculated as the sum of people
#that have the blood type divided by the population which is 10,000.
```

```r
Ofreq <- rep( c(0) , 10000 )
Afreq <- rep( c(0) , 10000 )
Bfreq <- rep( c(0) , 10000 )
ABfreq <- rep( c(0) , 10000 )
```

```r
for ( i in 1:10000 ) {
Ofreq[i] = mean( simdataO [1:i])
Afreq [ i ] = mean( simdataA [ 1 : i ] )
Bfreq [ i ] = mean( simdataB [ 1 : i ] )
ABfreq [ i ] = mean( simdataAB [ 1 : i ] )
}
```

```r
plot (1:1:1000 , Ofreq [ 1:1000 ] , type= 'l' , lty =1, ylim=c ( 0 , 1 ) )
lines ( 1 : 1 : 1000 , Afreq [ 1 : 1000 ] , col = 'red' , lty =2)
lines ( 1 : 1 : 1000 , Bfreq [ 1 : 1000 ] , col = 'green' , lty =3)
lines ( 1 : 1 : 1000 , ABfreq [ 1 : 1000 ] , col = 'blue' , lty =4)

#The lines are all becoming more consistent as the sum of the data measured increases increases, 
#they all start near each other where the mean is too small Due to the sample size being too small
# Once the sample size is large enough, the mean becomes constant as adding more people to the data won't affect as much.
```

```r
dice <- sample ( 1 : 6 , size = 7 , replace = TRUE)
sumDice <- sum( dice )
```

```r
event30 <- replicate ( 10000 , { dice <- sample ( 1 : 6 , size =7, replace= TRUE)
sum( dice )>30})

#trying to add intersection where  dices are rolled to have a sum between 30 and 35
event40 <- replicate ( 10000 , { dice <- sample ( 1 : 6 , size =7, replace= TRUE)
sum( dice )>30 } & { dice <- sample ( 1 : 6 , size =7, replace= TRUE)
sum( dice )<35 }) 
sum(event40)/10000

#trying to add intersection where  dices are rolled, and a die has to be one
event50 <- replicate ( 10000 , { dice <- sample ( 1 : 6 , size =6, replace= TRUE)
sum( dice )>30 } & { dice <- sample ( 1 , size =1, replace= TRUE)
sum( dice )<35 }) 
sum(event50)/10000

#Relative frequency for dices of sum between 30 and 35 was 0.0927, so 9% chance
#Relative frequency for dices are rolled, and a die has to be one is 0.8%, so a very low chance 
```

```r
eventfreq = rep(c(0) , 10000 )
for ( i in 1: 10000 ) {
eventfreq[i] = mean(event30[1:i])
}
plot ( 1: 10000 , eventfreq [ 1 : 10000 ] , type = 'l' )
mean( event30 )


# The plot seems to be going constant towards the mean, since the sample size increases the relative frequency 
#barely changes when  sample size increases a little.

```

```r
#Question 1

#Replicating for every event
eventdisks <- replicate ( 10000 , { Disks <- sample ( 1:5 , size =2, replace= FALSE)  
sum(Disks)==3}) 


mean(eventdisks)
#Relative Frequency that sum equals 3 is 0.1005

eventdisks2 <- replicate ( 10000 , { Disks <- sample ( 1:5 , size =2, replace= FALSE)
sum( Disks )>=6 & sum( Disks )<= 8  }) 
mean(eventdisks2)
#Relative Frequency that sum between 6 and 8 is 0.4913

#Repeating for first 1000 samples 
firstfreq <- rep( c(0) , 10000 )
secondfreq <- rep( c(0) , 10000 )

for ( i in 1:10000 ) {
firstfreq[i] = mean( eventdisks [1:i])
secondfreq [ i ] = mean( eventdisks2 [ 1 : i ] )

}

plot (1:1:1000 , firstfreq [ 1:1000 ] , type= 'l' , lty =1, ylim=c ( 0 , 1 ) )
lines ( 1 : 1 : 1000 , secondfreq [ 1 : 1000 ] , col = 'red' , lty =2)

#Possible solutions from10 outcomes is only 1+2, so the answer is 1/10, which is identical to the answer we have

#Possible solutions between 6 and 8 is 5, so 5/10 = 1/2 which is almost identical to the answer we have
```

```r
#Question 2

#1 is tails and 2 is Heads
# The probability of getting all tails in the first 9 tosses
Coinprop <- replicate ( 10000 , { Coins <- sample (1:2 , size =9, replace= TRUE)  
sum(Coins)==9}) 

#The probability of getting a head in the last toss
Coinprop2 <- replicate ( 10000 , { lastCoins <- sample (1:2 , size =1, replace= TRUE)  
sum(lastCoins)==2}) 
mean(Coinprop2) * mean(Coinprop)
# Answer is 0.00095285

#Theoriticaly, the probability of getting tails is (0.5)^9 * probability of getting head in the last, which is (0.5)
#So the answer is 0.00097 which is almost identical to the answer
```

```r
#Question 3

eventA <- replicate ( 10000 , { Dices <- sample ( 1:6 , size =2, replace= TRUE)  
sum(Dices)==3})

eventB <- replicate ( 10000 , { Dices2 <- sample ( 1:6 , size =2, replace= TRUE)  
sum(Dices2)==3 | sum(Dices2)==5}) 

 

#Calculating(P(A|B)), calculating the events that we want divided by all possible events. 0.21482236
((mean(eventA)*mean(eventB))+(mean(eventA)*(1-mean(eventB)))+((1-mean(eventA))*(mean(eventB))))/((mean(eventA)*mean(eventB))+(mean(eventA)*(1-mean(eventB)))+((1-mean(eventA))*(mean(eventB)))+((1-mean(eventA))*(1-mean(eventB))))
#Calculate The event of P(A|B) = (P(B|A)P(A))/P(B). Answer is 0.21, we can see that the answers are identical
```
