```r
library(ggplot2)
library(dplyr)
```

```r
Teeth=as_tibble(ToothGrowth)
```

```r
Teeth %>% ggplot(aes(x=dose, y = len)) +
  geom_point() +
  geom_smooth(method = "lm")
## last line puts a 95% confidence interval around our line, meaning if we were to repeat this experiment and
## fit a line, 95% of the time, it would be in the grey area
```

```r
Linear_Model = lm(len ~ dose, data=Teeth) #length is explained by dose
summary(Linear_Model) ## the 1.24e-14 is testing H_0: There is no linear relationship and H_1: There is a linear relationship
```

```r
cor(Teeth$len, Teeth$dose)
```

```r
predict(Linear_Model, data.frame(dose = .75), interval = "confidence")
#This gives you the predicted value for a dose based on your linear regression, and the assiciated confidence interval
```

```r
Teeth_Range = data.frame(dose = seq(.5,2,.5))

prediction_bounds <- predict(Linear_Model, newdata = Teeth_Range, interval = "p")
colnames(prediction_bounds) <- c("p_fit","p_lwr","p_upr")

prediction_bounds
bounds_data <- cbind(Teeth_Range, prediction_bounds)

tooth_plot <- Teeth %>% ggplot(aes(x = dose, y =len)) + 
  geom_point() + 
  geom_smooth(method = "lm")
tooth_plot <- tooth_plot +
  geom_line(data = bounds_data, aes(y = p_lwr), color="orange") +
  geom_line(data = bounds_data, aes(y = p_upr), color="orange")

tooth_plot #the orange line is our 95% confidence interval for our data, under the assumptions they are i.i.d and 
# their error is a normal distribution sith the same variance
```

```r
anova(Linear_Model)
```

```r
Teeth$supp <- factor(Teeth$supp)
Mult_Model = lm(len ~ supp + dose, data=Teeth)
summary(Mult_Model)
```

```r
pairwise.t.test(Teeth$len, Teeth$dose)
```

```r
pairwise.t.test(Teeth$len, Teeth$supp)
```

```r
ChickWeight = as_tibble(ChickWeight)
CW = ChickWeight %>% group_by(Chick) %>% mutate(avg_w = mean(weight)) 
ggplot(ChickWeight, aes(x = Diet, y = weight)) + 
  geom_boxplot()
ggplot(CW, aes(x = Diet, y = avg_w)) + 
  geom_boxplot()
#%>% select("Chick","Diet","avg_w")
```

```r
pairwise.t.test(ChickWeight$weight, ChickWeight$Diet)
pairwise.t.test(CW$avg_w, CW$Diet)
```
