```r
#Exponential
library(ggplot2)
data = read.csv("P2Data.csv")


hist(data$Data.set.A, freq = FALSE)
lines(density(data$Data.set.A))

mean(data$Data.set.A)
var(data$Data.set.A)

#It is exponential as the variance is almost the square of the mean, also because the graph is decreasing  subsequently
#Parameters 0<= x <3
```

```r
#Chi-Square

hist(data$Data.set.B, freq = FALSE)
lines(density(data$Data.set.B))
mean(data$Data.set.B)
var(data$Data.set.B)

# Chi-Square, as  the  variance is almost equal to the mean *2 
#Parameters 0<= x <16
```

```r
# Normal Uniform
d <- density(data$Data.set.C) # returns the density data
plot(d)

hist(data$Data.set.C, freq = FALSE)
lines(density(data$Data.set.C))

mean(data$Data.set.C)
var(data$Data.set.C)

#It is Normal due to being symmetric about the mean, 
#showing that data near the mean are more frequent in occurrence than data far from the mean.
#Parameters -inf<= x <inf
```
