```r
head(starwars)
Groups = as_tibble(starwars) %>%  
    group_by(sex,mass) %>%
    summarize(mean = mean(height)) 
Male = Groups %>% filter(sex == 'male')
Female = Groups %>% filter(sex == 'female')

Groups %>% filter(mass>50) 
order = arrange(starwars,desc(mass))
summarize(order, sex, mass )

mass = Groups$mass
mean = Groups$mean
ggplot(Groups, aes(x = mass, y = mean, size = 8, color = sex)) +  geom_point()

#The hypothesis that I predicted was that the majority of the weights of the male characters will be more than the female characters.
#Based on the data results seen below, we can see that the hypothesis was correct in which both in the data and the graph, male 
#characters were the majority in an increase of weight. Where 8 of the top 10 characters that had the highest mass were males.

```

```r
#Question 3
#My hypothesis is that male mean wight(M) is more than female weight( 55.31429), so H_1: M> 55.31429,
#the null hypothesis would be that H_0: M= 55.31429
t.test(Female$mass,mu = 0, alternative ='less', conf.level=.98)
t.test(Female$mass,mu = 0, alternative ='greater', conf.level=.98)
t.test(Male$mass, mu = 55.31429 , alternative = "greater", conf.level=.9)
#we reject H_0 at a confidence level of .9, #p-value 0.01>0.0001711

#My hypothesis is that Female mean height(M) is more than male height( 173.3083), so H_1: M> 173.3083, 
#the null hypothesis would be that H_0: M= 173.3083

t.test(Male$mean,mu = 0, alternative ='less', conf.level=.98)
t.test(Male$mean,mu = 0, alternative ='greater', conf.level=.98)

t.test(Female$mean, mu = 173.3083 , alternative = "greater", conf.level=.9)

#we fail to reject H_0 at a confidence level of .9, as 0.9 < p-value (0.9089) and, but for confidence 0.95 we can reject the null hypothesis 
t.test(Male$mean-Female$mean,mu = 0, alternative ='less', conf.level=.98)
t.test(Male$mean-Female$mean,mu = 0, alternative ='greater', conf.level=.98)

t.test(Male$mass-Female$mass, mu = 2.046296 , alternative = "greater", conf.level=.95)
#My hypothesis is that difference between male mean wight and female mean weight is more than the difference between heights(2.046296)
#so H_1: M> 2.046296,
#the null hypothesis would be that H_0: M= 2.046296

#We reject the null hypothesis as probabilities 0.05>0.001068 

#extra: testing normality for first t.test
ggplot(Female, aes(sample = Female$mean )) + stat_qq() +stat_qq_line()
#distribution is not normal meaning(as points do not lie on the line) we fail to reject the null hypothesis
```
