```python
f = open("hamlet.txt", "r")
```

```python
ftext = f.read()
```

```python
f.close()
```

```python
ftokens = ftext.split()
```

## Picking some words

Get every 2nd word between the 1st and the 3th: i.e. 1st, and 3th.

```python
print(ftokens[1:4:2])
```

```python
print(ftokens[1:4]+ftokens[1:4:2])
```

```python
fstring="-".join(ftokens[1:4:2])
```

```python
print(fstring)
```

```python
for i in range(len(ftokens)):
    if( len(ftokens[i])) == 20:
        print(ftokens[i])
```

```python
my_dict = {i:ftokens.count(i) for i in ftokens}

print (my_dict)
```

```python
import matplotlib.pyplot as plt

unique = []
names = []
for i in my_dict:
    if my_dict[i] >150:
        unique.append((my_dict[i]))
        names.append(i)
        
plt.bar(unique, names)
plt.title('Words Vs Count')
plt.show()
```
