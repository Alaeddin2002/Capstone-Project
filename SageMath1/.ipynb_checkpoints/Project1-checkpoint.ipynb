```python
#For Z2
A = matrix(Zmod(2), [[24,15,17,7], [3,19,20,8],[1,0,15,9]])
print (A.rref ())
print("")

# a = 1-c
# b = 1-c

#For Z3
A = matrix(Zmod(3), [[24,15,17,7], [3,19,20,8],[1,0,15,9]])
print (A.rref ())
print()

# a = 0
# b = 1
# c = 2

#For Z37
A = matrix(Zmod(37), [[24,15,17,7], [3,19,20,8],[1,0,15,9]])
print (A.rref ())
print("")

# a = -15c
# b = -24c
# c = 1

# So a = -15, b = -24

#For Q
A = matrix(QQ, [[24,15,17,7], [3,19,20,8],[1,0,15,9]])
print (A.rref ())
print("")

# a = -6/3071
# b = -646/3071
# c = 1843/3071
```
